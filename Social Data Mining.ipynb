### `Web Scrapping`

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

reviewlist = []


def get_soup(url):
    r = requests.get('https://www.amazon.com/product-reviews/B07ZPC9QD4/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&filterByStar=all_stars&reviewerType=all_reviews&pageNumber=1#reviews-filter-bar', params={'url': url, 'wait': 2})
    soup = BeautifulSoup(r.text, 'html.parser')
    return soup


def get_reviews(soup):
    reviews = soup.find_all('div', {'data-hook': 'review'})
    try:
        for item in reviews:
            review = {
                'product': soup.title.text.replace('Amazon.com: Customer reviews: ', '').strip(),
                'title': item.find('a', {'data-hook': 'review-title'}).text.strip(),
                'rating': float(
                    item.find('i', {'data-hook': 'review-star-rating'}).text.replace('out of 5 stars', '').strip()),
                'review': item.find('span', {'data-hook': 'review-body'}).text.strip(),
            }
            reviewlist.append(review)
    except:
        pass


for x in range(1, 200):
    soup = get_soup(
        f'https://www.amazon.com/product-reviews/B07ZPC9QD4/ref=cm_cr_arp_d_viewopt_sr?ie=UTF8&filterByStar=all_stars&reviewerType=all_reviews&pageNumber=1#reviews-filter-bar={x}')
#     print(f'Getting page: {x}')
    get_reviews(soup)
#     print(len(reviewlist))
    if not soup.find('li', {'class': 'a-disabled a-last'}):
        pass
    else:
        break

df = pd.DataFrame(reviewlist)
df.to_csv(r'airpod_pro.csv', index=False)
```

### `Merging Dataset`

### `Data Cleaning`

```python
import typing
from nltk.tokenize import word_tokenize
import nltk
from nltk.corpus import stopwords
import string
from string import digits
import numpy as np


# data cleaing
stop = stopwords.words('english')

# lower cases
full_dataset['lower_summary'] = full_dataset['review'].apply(lambda x : x.lower())

# stop word
full_dataset['clear_summary'] = full_dataset['lower_summary'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))

# removing duplicates
full_dataset['clear_summary'].drop_duplicates()

# removing punctuations
full_dataset['clean_summary'] = full_dataset['clear_summary'].apply(lambda x:''.join([i for i in x if i not in string.punctuation]))

full_dataset['clean_summary'] = full_dataset['clean_summary'].str.replace('[^\w]', '', regex = False)

# removing numbers
full_dataset['clean_summary'] = np.where(full_dataset.clean_summary.str.contains('^\d+$'), full_dataset.clean_summary, full_dataset.clean_summary.str.replace('\d+', ''))


# tokenizing 
full_dataset['tokenized_summary'] = full_dataset.apply(lambda row: nltk.word_tokenize(row['clean_summary']), axis=1)

full_dataset
```

### `Sentiment Analysis`

```python
from textblob import TextBlob
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer


# Sentiment score from TextBlob
def get_textblob_sentiment(text):
    analysis = TextBlob(text)
    return analysis.sentiment.polarity

# Sentiment score from Vader
def get_vader_sentiment(txt):
    analyzer = SentimentIntensityAnalyzer()
    vs = analyzer.polarity_scores(txt)
    return vs['compound']

full_dataset['sentiment'] = full_dataset['clean_summary'].apply(lambda x : (get_vader_sentiment(x) + get_textblob_sentiment(x))/2)


full_dataset
```

```python
full_dataset['sentiment_status'] = full_dataset['sentiment'].apply(lambda c: 'neutral' if c==0 else 'positive' if c>=-0.01 else 'negative')
full_dataset
```
